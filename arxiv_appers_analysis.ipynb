```python
import pandas as pd
import arxivscraper
import math
import matplotlib as mpl
from calendar import monthrange
from matplotlib import pyplot as plt
import os
```

1 Data pull from arXiv
====

```python
def pull_data (category, date_from, date_until):
    scraper = arxivscraper.Scraper(category=category, date_from=date_from,date_until=date_until)
    output = scraper.scrape()
    df = pd.DataFrame([])
    if (output!=1):
        cols = ('id', 'title', 'categories', 'abstract', 'doi', 'created', 'updated', 'authors')
        df = pd.DataFrame(output,columns=cols)
    return(df)
```

```python
cats = ['stat', 'cs']
years = list(range(2007, 2020))
for category in cats:
    for year in years:
        print(f"\n=====\nQuerying metadata in arXiv category {category}, year {year}\n=====\n")
        df = pd.DataFrame([])
        for month in list(range(1, 13)):
            month_digits = str(month).zfill(2)
            lastday = monthrange(year, month)[1]
            df0 = pull_data(category, f"{year}-{month_digits}-01", f"{year}-{month_digits}-{lastday}")
            df = pd.concat([df, df0])
        df.to_csv(f"arxiv_metadata_{category}_{year}.csv", index=False)
        numrecords = format(len(df), ',')
        print(f"\nSaved {numrecords} total records for {year}, {category} category\n")

print('done')
```

```python
sdf = pd.concat([pd.read_csv(ff) for ff in list(filter(lambda x: '_stat_' in x, os.listdir('.')))])
cdf = pd.concat([pd.read_csv(ff) for ff in list(filter(lambda x: '_cs_' in x, os.listdir('.')))])
```

2 Exploratory analysis of selected stat and cs categories for code availability
===

```python
def findcategory (category, df):
    return df.categories.apply(lambda x: category in x)
def findcategories (categories, df):
    return df.categories.apply(lambda x: bool(sum([category in x for category in categories])))
def findterm (term, df):
    return df.abstract.apply(lambda x: term in x)
def findterms (terms, df):
    return df.abstract.apply(lambda x: bool(sum([term in x for term in terms])))
```

```python
categories = ['stat.th', 'stat.ap', 'stat.ml']
cscategories = ['cs.cv', 'cs.ai']
findcategories(categories + cscategories, pd.concat([sdf, cdf])).sum()
```

```python
def categorytermcontent (category, terms, df):
    newdf = df[findcategory(category, df)]
    mark = findterms(terms, newdf)
    count = int(mark.sum())
    total = len(newdf)
    pct = 0
    se = 0
    if total > 0:
        pct = float(count) / float(total)
        se = math.sqrt(pct * (1 - pct) / total)
    return {'category' : category, 'terms' : terms, 'count' : count, \
            'percent' : pct, 'total' : total, \
           'se' : se}
```

```python
def explore_terms (terms, categories, df):
    for term in terms:
        print(f"\n====\nCounts of abstracts with term '{term}':\n====\n")
        for category in categories:
            res = categorytermcontent(category, [term], df)
            print(f"{res['category']}: {res['count']}")
    return(0)
```

```python
terms = ['github', 'sourceforge', 'open source', 'code', 'r package', 'python module', 'python package', 'r module']
explore_terms(terms, categories, sdf)
```

```python
cscategories = ['cs.cv', 'cs.ai']
explore_terms(terms, cscategories, cdf)
```

3 Charts of code availability trends over time
====

```python
def create_count_time_series (categories, terms, years, df):
    result = []
    for category in categories:
        for term in terms:
            for year in years:
                yearmark = df.created.apply(lambda x: str(year) in x)
                df1 = df[yearmark]
                res = categorytermcontent(category, [term], df1)
                res['year'] = year
                result.append(res)
    return(pd.DataFrame(result))

def create_count_trends (categories, terms, years, df):
    result = []
    for category in categories:
        for year in years:
            yearmark = df.created.apply(lambda x: str(year) in x)
            df1 = df[yearmark]
            res = categorytermcontent(category, terms, df1)
            res['year'] = year
            result.append(res)
    return(pd.DataFrame(result))
```

```python
counts = create_count_time_series(categories, terms, years, sdf)
```

```python
ct = create_count_trends(categories, terms, years, sdf)
cst = create_count_trends(cscategories, terms, years, cdf)
```

```python
def plot_trends (trends, filename='trends.png'):
    fig, ax = plt.subplots(nrows=2, figsize=(9,6))
    for name, group in trends.groupby(['category']):
        group.plot(x='year', y='percent', ax=ax[0], label=name + ' % articles')
        ax[0].fill_between(group.year, group.percent - group.se, group.percent + group.se, alpha=0.35)

    l = ax[0].get_lines()

    count = 0

    for name, group in trends.groupby(['category']):
        group.plot(x='year', y='total', ax=ax[1], label=name + ' # articles', style='--',\
                  color=list(i.get_color() for i in l)[count])
        count+=1

    #ax[0].yaxis.grid(which="major")
    #ax[1].yaxis.grid(which="major")
    
    ax[0].set_ylabel('arXiv articles using one of terms')
    ax[1].set_ylabel('arXiv number of articles')

    vals = ax[0].get_yticks()
    ax[0].set_yticklabels(['{:,.0%}'.format(x) for x in vals])
    
    ax[1].get_yaxis().set_major_formatter(mpl.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
    ax[0].legend(loc='upper left')
    ax[1].legend(loc='upper left')
    plt.savefig(filename, dpi=500, bbox_inches='tight')
```

```python
plot_trends(ct, filename='stat_trends2007_to_2019.png')
```

```python
plot_trends(cst, filename='cs_trends2007_to_2019.png')
```

```python
ctc = create_count_trends(categories + cscategories, terms, years, pd.concat([sdf, cdf]))
```

```python
plot_trends(ctc, filename='combined_trends_2007_to_2019.png')
```
